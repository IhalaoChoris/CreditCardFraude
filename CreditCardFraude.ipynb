```python
from sklearn.preprocessing import StandardScaler, normalize

#aqui iremos pré-processar os dados para que eles sejam categóricos

#df_10x.drop('Time', axis=1, inplace=True) ------> por algum motivo, algumas funções, como essa, impedem o treinamento de modelos de classificação, pois estavam impedindo a normalização dos vetores
df_10x.iloc[:, :-1] = StandardScaler().fit_transform(df_10x.iloc[:, :-1])
#df_10x = StandardScaler().fit_transform(df_10x)
df_10x

x = df_10x.iloc[:, :-1].values
y = df_10x.iloc[:, -1].values

x = normalize(x, norm='l1')

print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42, stratify=y)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.utils.class_weight import compute_sample_weight
import time

w_train = compute_sample_weight('balanced', y_train) #quando não utilizado, a acurácia subiu de 97% para 99%

dec_tree = DecisionTreeClassifier(max_depth = 4, random_state = 35)


time_zero = time.time()
dec_tree.fit(x_train, y_train)
training_time = time.time() - time_zero
print("Training time: ", training_time)
```

```python


predictions = dec_tree.predict(x_test)

#validação

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("Accuracy: ", accuracy_score(y_test, predictions))
print("Confusion Matrix: \n", confusion_matrix(y_test, predictions))
print("Classification Report: \n", classification_report(y_test, predictions))
```

```python
#aplicando o método Support Vector Machine
from sklearn.svm import LinearSVC

svm = LinearSVC(random_state=35, loss='hinge', fit_intercept=False, max_iter=5000)

time_zero = time.time()
svm.fit(x_train, y_train)
training_time = time.time() - time_zero
print("Training time: ", training_time)
```

```python
#validação

print("Accuracy: ", accuracy_score(y_test, svm.predict(x_test)))
print("Confusion Matrix: \n", confusion_matrix(y_test, svm.predict(x_test)))
print("Classification Report: \n", classification_report(y_test, svm.predict(x_test)))

#notas: Não há um bom resultado para os FN, é possivel notar pelos resultados
# de validação. Porém, houve uma melhora significativa ao retirar o parametro
#class_weight='balanced', a acuracia dos Negativos subiu de 0.03 para 0.66
#necessário estudar mais profundamente como cada parametro afeta o treinamento
#para cada dataset diferente.
```
